# Create your first spreadsheet file

Assume you want to create a spreadsheet with the following data.

```Python
data = [
    ['James', 200],
    ['Jones', 250],
    ['Jasper', 125],
    ['Jason', 175]
]
```
...and calculate the sum inside the spreadsheet.

To work the the `xlsxwriter` framework you need to import it.

```Python
import xlsxwriter
```

Then you open a `workbook` with the filename and start a `worksheet` that you can name.

```Python
workbook = xlsxwriter.Workbook('example-01.xlsx')
worksheet = workbook.add_worksheet('My data')
```

To write to cells you use indexing and not the cell names `A1`, `A2`, ..., `B2`, `B2`, ...

|   | A | B | C | D |
|---|:---:|:---:|:---:|:---:|
| 1 | row: 0, col:0 | row: 0, col:1  | row: 0, col:2  | row: 0, col:3  |
| 2 | row: 1, col:0 | row: 1, col:1  | row: 1, col:2  | row: 1, col:3|
| 3 | row: 2, col:0 | row: 2, col:1  | row: 2, col:2  | row: 2, col:3|

You write to the worksheet with the `write()` method.

```Python
worksheet.write(row, col, cell_data)
```

You can write formulas.

```Python
worksheet.write(row, col, '=SUM(B1:B4)')
```

Finally, you need to close the workbook with `close()` which will write the data.

```Python
workbook.close()
```

````python
import xlsxwriter

workbook = xlsxwriter.Workbook('example-01.xlsx')
worksheet = workbook.add_worksheet('My data')

data = [
    ['James', 200],
    ['Jones', 250],
    ['Jasper', 125],
    ['Jason', 175]
]

row = 0
col = 0

for name, cost in data:
    worksheet.write(row, col, name)
    worksheet.write(row, col + 1, cost)
    row += 1
    
worksheet.write(row, 0, 'Total')
worksheet.write(row, 1, '=SUM(B1:B4)')

workbook.close()
````
