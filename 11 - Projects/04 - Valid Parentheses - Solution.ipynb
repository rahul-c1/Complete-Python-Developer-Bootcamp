# Valid Parentheses

A valid parentheses string satisfies the following:

- Every opening bracket must have a matching closing bracket of the same type.
- The brackets should be closed in the correct order.

Examples of valid
- `()`
- `(())`
- `()()`
- `(()())`

Examples of invalid
- `(()`
- `)(`
- `())(`
- `((())`

### Advanced
Can you solve it if the string contains simple, curly, and square braces: `()[]{}`

Examples of valid
- `()[]`
- `([])`
- `(){}`
- `([]{})`

Examples of invalid
- `([)]`
- `(}(]`
- `()(}`
- `({][})`

```python
def matching_parentheses(s):
    no_open = 0
    
    for char in s:
        if char == ')':
            no_open -= 1
        elif char == '(':
            no_open += 1
            
        if no_open < 0:
            return False
        
    if no_open > 0:
        return False
    
    else:
        return True


assert matching_parentheses('()')
assert matching_parentheses('(())')
assert matching_parentheses('()()')
assert matching_parentheses('(()())')
assert matching_parentheses('(') == False
assert matching_parentheses(')(())') == False
assert matching_parentheses('(()') == False
assert matching_parentheses('(()()') == False
```

```python
def matching_parentheses(s):
    stack = []
    
    for char in s:  # s = ()[] # char = ')'
        if char in '([{':
            stack.append(char)  # stack = []
            
        if char in ')]}':
            if len(stack) == 0:
                return False
            
            top = stack.pop()  # top = '('
            
            if top == '(' and char != ')':
                return False
            if top == '[' and char != ']':
                return False
            if top == '{' and char != '}':
                return False
            
    if len(stack) != 0:
        return False
    else:
        return True
    

assert matching_parentheses('(') == False
assert matching_parentheses('()[]')
assert matching_parentheses('([])')
assert matching_parentheses('(){}')
assert matching_parentheses('([]{})')
assert matching_parentheses('([)]') == False
assert matching_parentheses('(}(]') == False
assert matching_parentheses('()(}') == False
assert matching_parentheses('({][})') == False
```
