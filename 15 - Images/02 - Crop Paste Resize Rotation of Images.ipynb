# Crop, Paste, Resize, and Rotation of Images

### Crop

You can crop (get part of the image) by applying the `crop()` method on the image object.

```Python
from PIL import Image

laptop = Image.open('laptop.jpg')

laptop.crop((0, 0, 100, 100))
```

### Paste

```Python
crop = laptop.crop((200, 150, 300, 250))

laptop.paste(im=crop, box=(0, 0))
```

### Resize

You can resize images

```Python
h, w = laptop.size

h_new = h//3
w_new = w//3

laptop.resize((h_new, w_new))
```

You can also stretch.

````python
h, w = laptop.size
````

````python
h_new = h//3
w_new = w//3

laptop.resize((h_new, w_new))
````
