# Pillow library

You can read images and get a lot of information about it as follows.

```Python
from PIL import Image

laptop = Image.open('laptop.jpg')

# Data type
type(laptop)

# Size / dimensions
laptop.size

# Filename
laptop.filename

# Format encoding
laptop.format_description
```

You can view the image as follows in a Notebook.

```Python
laptop
```

If you use an IDE, you need to call `.show()`

```Python
laptop.show()
```
